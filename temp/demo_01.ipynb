```python
import qlib
from qlib.constant import REG_CN
provider_uri = "/Users/liupeng/.qlib/crypto_data/ql/"  # target_dir
qlib.init(provider_uri=provider_uri, region=REG_CN)
```

```python
from qlib.data import D
D.calendar(start_time='2025-10-01', end_time='2025-10-05', freq='15min')
```

```python

instruments = D.instruments(market='all')
D.list_instruments(instruments=instruments, start_time='2025-10-01', end_time='2025-10-05', as_list=True)[:6]
```

```python
fields = ['date','open','high','low','close','volume','symbol']
D.features(instruments, fields, start_time='2023-01-01', end_time='2023-01-02', freq='day').head().to_string()
```
